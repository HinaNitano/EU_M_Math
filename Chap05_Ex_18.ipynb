```python
import numpy as np
import scipy as sp

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

%precision 3
```

```python
from scipy.optimize import fsolve
```

```python
def f(x):
  y = x**3 -2*x**2 - 11*x + 12
  return y
```

```python
x = np.linspace(-4, 4)
plt.plot(x, f(x))
plt.plot(x, np.zeros(len(x)))
plt.grid(True)
```

```python
x = fsolve(f, 0)
print(x)
```
