```python
import numpy as np
import math

# シード値の固定
np.random.seed(0)

# 乱数の発生
x = np.random.uniform(0, 1, 10000)
y = np.random.uniform(0, 1, 10000)

# 円の内部に入る点のカウント
count_inside = 0
for i in range(len(x)):
  # 点(x[i], y[i])が円の内部に入るかどうかを判定
  if math.hypot(x[i], y[i]) <= 1:
    count_inside += 1

# 結果の表示
print(f'円の内部に入る点の個数: {count_inside}')
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import math

# 円の内部の点か外部の点か判別
def plot_points_inside_outside_circle(num_samples, seed=None):
    if seed is not None:
        np.random.seed(seed)
    points = np.random.uniform(0, 1, size=(num_samples, 2))
    inside_points_x = []
    inside_points_y = []
    outside_points_x = []
    outside_points_y = []
    for point in points:
        if math.hypot(point[0], point[1]) < 1:
            inside_points_x.append(point[0])
            inside_points_y.append(point[1])
        else:
            outside_points_x.append(point[0])
            outside_points_y.append(point[1])

# 散布図
    plt.figure(figsize=(8, 8))
    plt.scatter(inside_points_x, inside_points_y, color='blue', label='Inside Circle')
    plt.scatter(outside_points_x, outside_points_y, color='red', label='Outside Circle')
    circle = plt.Circle((0, 0), 1, color='black', fill=False)
    plt.gca().add_patch(circle)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Points Inside and Outside Circle')
    plt.legend()
    plt.show()

num_samples = 10000
plot_points_inside_outside_circle(num_samples)
```
