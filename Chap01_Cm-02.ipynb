```python
def is_prime(num):
# 引数で与えられた数が素数かどうかを判定する関数

# numが素数であればTrue、そうでなければFalse
  if num <= 1:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

def primes_up_to_N(N):
# 自然数Nまでの素数を表示する関数
    primes = [i for i in range(2, N+1) if is_prime(i)]
    return primes

N = int(input("自然数Nを入力してください: "))
prime_list = primes_up_to_N(N)
print(f'{N}までの素数は:', prime_list)
```
